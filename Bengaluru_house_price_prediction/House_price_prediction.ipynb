* One can clearly see that there is a value "1133-1384" in the 'total-sqft' column. It's a range and we need to convert it into a single numeric value.

```python
def convert_range(x):
    temp=x.split('-')
    if len(temp)==2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df3['total_sqft']=df3['total_sqft'].apply(convert_range)
```

* In the above code we make a user defined function "convert_range", temp = x.split('-'). It splits the string in to two numbers, like '1133-1384' to ['1133', '1384'], then if the length == 2 then it calculates the average of those two values.

# Outlier detection and removing 

```python
df4= df3[((df3["total_sqft"]/df3["bhk"])>=300)]
df4.shape
```

* Here we remove the rows, where the average area of each bedroom is less than 300 square feet.

* Here we got a minimum price_per_sqft of 267.82 and a maximum price_per_sqft of 176470.58 however a mean price_per_sqft of 6308.50 .

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame() # Initialize the empty DataFrame
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft) # mean
        st= np.std(subdf.price_per_sqft) # Standard deviation

        gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_output = pd.concat([df_output,gen_df],ignore_index = True)
    return df_output
df5=remove_outliers_sqft(df4)
df5.shape
```

* here we used a user defining function(remove_outliers_sqft) for outlier treatment in price_per_sqft column.

```python
df5.head()
```

```python
def bhk_outlier_remover(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LinearRegression, Lasso , Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
import pandas as pd

def find_best_model_using_gridsearchcv(X, y):
    # Scale the data if needed
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    algo = {
        'Linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]  # Removed 'normalize'
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'ridge': {
            'model': Ridge(),
            'params': {
                'alpha': [1,2]
            }
        },
        'Decision_tree':{
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],  # Updated 'mse' to 'squared_error'
                'splitter': ['best', 'random'] 
            }
        }
        
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    
    for algo_name, config in algo.items():
        gs = GridSearchCV(config['model'], config['params'], cv = cv, return_train_score=False)
        gs.fit(X_scaled, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
find_best_model_using_gridsearchcv(X, y)
```

```python
def predict_price(location, sqft, bath, bhk):
    # Get the index of the location column in the feature matrix
    loc_index = np.where(X.columns == location)[0][0]

    # Create a feature vector initialized with zeros
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk

    # Set the location feature to 1 based on the location index
    if loc_index >= 0:
        x[loc_index] = 1

    # Convert the feature vector to a DataFrame with column names
    x = pd.DataFrame([x], columns=X.columns)  # This ensures the correct column names

    # Make the prediction using the model
    return lr_model.predict(x)[0]
```
